```

from IPython.utils import encoding
### TextLoader

from langchain_community.document_loaders import TextLoader

loader = TextLoader("./data/text_files/python_intro.txt", encoding="utf-8")
document = loader.load()
print(document)
```

```
### Directory Loader
### load text files
from langchain_community.document_loaders import DirectoryLoader

loader = DirectoryLoader("./data/text_files/", glob="**/*.txt", loader_cls=TextLoader, loader_kwargs={'encoding':'utf-8'}, show_progress=True)
documents = loader.load()
print(documents)
```

```
# Load PDF files
from langchain_community.document_loaders import DirectoryLoader, PyPDFLoader, PyMuPDFLoader


# Create a dummy PDF file for demonstration if needed, or assume existing ones.
# For this example, let's assume there are PDF files in the directory.

# Initialize DirectoryLoader to load PDF files
loader = DirectoryLoader(
    "./data/pdf/",
    glob="**/*.pdf", # Target PDF files
    #loader_cls=PyPDFLoader, # Use PyPDFLoader for PDF files
    loader_cls=PyMuPDFLoader, # Use PyMuPDFLoader for PDF files
    loader_kwargs={'extract_images': True}, # Extract images from PDF
    show_progress=True
)

# Load the documents
pdf_documents = loader.load()

# Print the loaded documents
print(pdf_documents)
```
